```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
data=pd.read_csv("kc_house_data.csv")
data.head()
```

```python
price=data['price'].values
grade=data['grade'].values
floors=data['floors'].values
bedrooms=data['bedrooms'].values
sqft_living=data['sqft_living'].values
```

```python
from mpl_toolkits.mplot3d import Axes3D
fig = plt.figure()
ax = Axes3D(fig)
ax.scatter(price,bedrooms,sqft_living,floors,bedrooms, color="#ef1234")
plt.show()
```

```python
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error

x=np.array([grade,sqft_living,floors,bedrooms]).T
y=np.array(price)
```

```python
reg=LinearRegression()
reg=reg.fit(x,y)
y_pred=reg.predict(x)

rmse=np.sqrt(mean_squared_error(y,y_pred))
r2=reg.score(x,y)
print(rmse)
print(r2)
```

```python
print(reg.coef_)
print(reg.intercept_)
```
